```python
import yfinance as yf #yahoo finance api
nf500t = yf.Ticker("^CRSLDX") #ticker symbol of nifty 500
nf500h = nf500t.history(period="max")
```

```python
nf500h.plot.line(y="Close", use_index=True)
```

```python
del nf500h["Dividends"]
del nf500h["Stock Splits"]
# delete columns dividend and stock splits
```

```python
nf500h["NextDay"] = nf500h["Close"].shift(-1) #create new column called next day and use the closing price of next day
```

```python
nf500h
```

```python
nf500h["Target"] = (nf500h["NextDay"] > nf500h["Close"]).astype(int) #create another column for target feature, if next day closing higher than prev day closing then 1, or else 0
nf500h
```

```python
nf500h.index
```

```python
nf500h = nf500h.loc["2011-01-01":].copy() # copy only data after 2011
nf500h
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
#n_estimators is the number of random forest trees, min_samples_split helps to reduce overfitting, random_state=1 mentions the randomness of each tree to be the same randomness
#so that the next time you run it, it takes the same random parameters and you get similar testinf and training results

train = nf500h.iloc[:-100]
test = nf500h.iloc[-100:]


#now we train
predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score #precision score is exactly what it sounds like
import pandas as pd


#prediction
preds = model.predict(test[predictors])# returns np array, therefore convert to pandas sereis
preds =pd.Series(preds, index=test.index)# use same index as previous dataset
preds
```

```python
#findinf accuracy
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

```python
#backtesting system
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
  all_predictions = []
  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train, test, predictors, model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

```python
predictions = backtest(nf500h, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []
for horizon in horizons:
  rolling_averages = nf500h.rolling(horizon).mean()

  ratio_column = f"Close_Ratio_{horizon}"
  nf500h[ratio_column] = nf500h["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  nf500h[trend_column] = nf500h.shift(1).rolling(horizon).sum()["Target"]

  new_predictors+= [ratio_column, trend_column]
```

```python

def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1] #returns probablility instead of 1 or 0
  preds[preds >= .6] = 1
  preds[preds < .6] = 0       #increasing the confidence of the model
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
predictions = backtest(nf500h, model, new_predictors)
```
